```python
## Task-4: Prepare data before training the model
cancer_df
```

```python
target_df = cancer_df['Biopsy']
input_df = cancer_df.drop(columns = ['Biopsy'])
```

```python
target_df.shape
```

```python
input_df.shape
```

```python
X = np.array(input_df).astype('float32')
y = np.array(target_df).astype('float32')
```

```python
y.shape
```

```python
# Scaling the data before feeding the model
from sklearn.preprocessing import StandardScaler, MinMaxScaler
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
X
```

```python
# spliting the data in to test and test sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
X_test, X_val, y_test, y_val = train_test_split(X_test, y_test, test_size = 0.5)
```

```python
## Task-5: Train and Evaluate XGBOOST Classifier
!pip install XGBOOST
```

```python
# Train an XgBoost classifier model
import xgboost as xgb

model = xgb.XGBClassifier(learning_rate = 0.1, max_depth = 50, n_estimators = 100)

model.fit(X_train, y_train)
```

```python
result_train = model.score(X_train, y_train)
result_train
```

```python
# predict the score of the trained model using the testing dataset
result_test = model.score(X_test, y_test)
result_test
```

```python
# Make predictions on the test data
y_predict = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
print(classification_report(y_test, y_predict))
```

```python
cm = confusion_matrix(y_predict, y_test)
sns.heatmap(cm, annot = True)
```
